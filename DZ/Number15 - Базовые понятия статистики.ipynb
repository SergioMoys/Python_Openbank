```python
data = pd.read_csv('horse_data.csv', header = None, na_values = '?')
df = data[[(1),(2),(4),(5),(6),(7),(11),(23)]]
df.columns =  ['surgery', 'Age', 'rectal temperature', 'pulse', 'respiratory rate', 'temperature of extremities', 'pain', 'outcome']
df
```

### Задание 2. Первичное изучение данных

```python
#Выброс

print(df['pulse'].median())
print(df[(df['pulse'] >10) & (df['pulse'] < 30)]['pulse'].median())
```

```python
#Выбросы as Lection

q1 = df.pulse.quantile(0.25)
q3 = df.pulse.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df['pulse'].between(lower_bound, upper_bound, inclusive=True)].sort_values('respiratory rate')
remove_outliers
```

Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе заполнения пропусков по каждому столбцу на основе рассчитанных статистик и возможной взаимосвязи значений в них. Сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
df['pain'].fillna('0', inplace=True)
print(df['pain'].isna().sum())
print(df['pain'].unique())
```

```python
df['rectal temperature'].fillna(df['rectal temperature'].mean, inplace=True)
```

```python
df['temperature of extremities'].fillna(df['temperature of extremities'].median, inplace=True)
print(df['temperature of extremities'].isna().sum())
print(df['temperature of extremities'].describe())
```

```python
df['respiratory rate'].fillna(df['respiratory rate'].mode, inplace=True)
print(df['respiratory rate'].isna().sum())
print(df['respiratory rate'].describe())
```

```python
df['pulse'].fillna(df['pulse'].median, inplace=True)
print(df['pulse'].isna().sum())
print(df['pulse'].describe())
```
